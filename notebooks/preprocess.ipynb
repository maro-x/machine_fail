```python
data = pd.read_csv("data/data.csv")
```

```python
data.isna().sum()
```

```python
num_col = ['footfall','AQ','USS','CS','VOC','RP',"IP","Temperature"]
cat_col = ['tempMode']
target = ['fail']
```

```python
x = data.drop(columns = target)
y = data[target]
```

```python
num_data = x[num_col]
cat_data = x[cat_col]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_numdat= pd.DataFrame(scaler.fit_transform(num_data),columns=num_col)
```

```python
import pickle
with open('data/scaler.pkl','wb') as f:
    pickle.dump(scaler,f)
```

```python
data_processed = pd.concat([scaled_numdat,cat_data,y],axis=1)
```

```python
data_processed
```

```python
data_processed.to_csv('data/preprocessed.csv',index=False)
```
